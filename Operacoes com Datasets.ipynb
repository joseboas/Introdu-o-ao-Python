**Divindo o dataset em colunas**

```python
ds = open('/content/clima.csv','r')
```

```python
data = ds.read()
```

```python
#Separa o arquivo com o comando split co  critério \n ou enter
linhas = data.split('\n')
```

```python
print(linhas)
```

```python
#Cria uma lista vazia
fullData = []
```

```python
#Percorro todas as linhas e colunas. A coluna pe separada por ',', mas poderia ser outro. Depende do seu arquivo.
for linha in linhas:
  split_linhas =  linha.split(',')
  fullData.append(split_linhas)
```

```python
print(fullData)
```

**Contando a quantidade de linhas de um arquivo**

**Contando a quantidade de colunas de um arquivo**

```python
arq = open('/content/clima.csv','r')
data = arq.read()
linha = data.split('\n')
fullData = []
```

```python
for linha in linhas:
  splitLinha = linha.split(',')
  fullData.append(splitLinha)  
  primLinha =  fullData[0]
  #dou um break, pois precisa apenas da primeira linha para contar as colunas
  break
```

```python
count = 0
for coluna in primLinha:
  if  coluna != '':
    count += 1
```

```python
print(count)
```
