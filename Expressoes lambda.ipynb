**As Expressões Lambdas sempre começam com a palavra reservada lambda**

```python
#quadrado é o resultado
#lambda palavra reservada
#primeiro x parê de entrada 
#x**2 é o retorna da função, expressão
quadrado =   lambda x:x**2
```

```python
quadrado(3)
```

```python
quadrado(5)
```

```python
#Verificar se um número é par
par = lambda x:x%2==0 
```

```python
par(4)
```

```python
par(5)
```

```python
#Primeiro elemento de uma string
primeiro = lambda x: x[0] 
```

```python
primeiro('Estudando Python')
```

```python
#Reverso
reverso = lambda x: x[::-1]
```

```python
reverso('Estudando Python')
```

```python
#Percorrendo uma string
percorre = lambda s:s[:]
```

```python
percorre('Estudando Python')
```

```python
#Percorrendo uma string até o último elemento
percorreAte = lambda s:s[:-1]
```

```python
percorreAte('Estudando Python')
```

```python
#soma dois numeros
soma = lambda x,y:x+y
```

```python
soma(5,3)
```
