**Indexando uma String**

```
#Python começa a indexação (índice) na posição 0
c = "José Luiz Vilas Boas"       
c
```

```
c[0]
```

```
c[1]
```

```
c[2]
```

**Percorrer uma string ou slicing com o :**

```
# A partir da posição 1, pega tudo
c[1:]
```

```
# A string permanece inalterada
c
```

```
#Vai o inicio (0) até a posição 5.
c[:6]
```

```
#Imprime tudo
c[:]
```

```
#Inicio do 0 e vou até o 3.
c[0:4]
```

```
#Ler de trás para frente com indexação negativa
c[-1]
```

```
#Leio a partir do -1 e vou até o final, mas o ultimo caractere não entra.
c[:-1]
```

**Podemos fatiar a string** 

```
#Vou ler toda a string pulando 1 posição
c[::1]
```

```
#Vou ler toda a string pulando 2 posições
c[::2]
```

```
#Pega de trás para frente de 1 em 1
c[::-1]
```

**Contando quantidade de caracteres encontrados na string String**
