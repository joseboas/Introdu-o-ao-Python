**Note: Tuplas são usadas quando não queremos que determinado conjunto de valores não sejam alterados, pois tupla é imutável.**

**Exemplos de Tuplas**

**Criando uma tupla**

```
t1 = ('José','Maria','Pedro',33)
t1
```

**Não tem como dar append em tuplas**

```
t1.append('Lucas')
```

**Não tem como exluir um elemento de uma tupla**

```
 t1.remove('Lucas') 
```

**Excluindo todos os elementos de uma tupla**

```
del(t1)
print(t1)
```

**Excluindo toda a tupla**

```
del t1
```

```
print(t1)
```

**Tupla dentro de outra tupla**

```
t2 = (1,2,(3,4,5,'False'),6,7,True)
t2
```

**Criando uma tupla com range**

```
t3 = tuple(range(0,10))
t3
```

**Obtendo o comprimento de uma lista**

```
print(len(t3))
```

**Acessando elementos**

```
print(t3[2])
```

**Slicing em uma tupla**

```
t3[1:]
```

```
t3[0:5]
```

**Verificando o índice**

```
t2 = ('José','Maria','Pedro',33)
t2.index('José')
```

```
t3[:-1]
```

**Outra forma de exibir tuplas**

```
t3[0:len(t3)]
```

```
t3[0:5]
```

**Percorrendo uma tupla**

```
for i in range(0,len(t3)):
  print(t3[i])
```

```
for elemento in t3:
  print(elemento)
```

**Como inserir um item na tupla?**

```
#convertendo uma tupla para lista
lst1 = list(t3)
```

```
lst1
```

```
#inserimos o elemento
lst1.append(10)
```

```
lst1
```

```
#convertermos novamente para tupla
tupla = tuple(lst1)
```

```
tupla
```
