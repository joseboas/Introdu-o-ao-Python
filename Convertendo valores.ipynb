**Conversão de valores**

**String para inteiro e string para float**

```
x = int(input("Informe um valor:"))
y = int(input("Informe outro valor:"))
print(f' A soma das variáveis = {x+y}')
```

```
z = float(input("Informe um valor:"))
c = float(input("Informe outro valor:"))
print(f' A soma das variáveis = {z+c}')
```

**int para string e float para string**

```
a = str(x)
b = str(y)
print(f'A soma das string são = {a+b}')
```

```
e = str(z)
f = str(c)
print(f'A soma das string são = {e+f}')
```

**float para int e int para float**

```
#Float to Int
g = int(z)
print(f' valor de g = {g}')
#Int to Float
h = float(z)
print(f' valor de h = {h}')
```

```
#Aqui não está arredondando, está convertente 7.5 para 7
i = 7.5
int(i)
```
