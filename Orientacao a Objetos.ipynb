**Exemplo da Classe Usuário 1º Exemplo**

**Pela sintaxe por convenção usamos o comando class**

```
#Criando uma classe Usuário.
class Usuario():
  #Criando um método para inicializar o objeto. O nome do método é __init__.
  #Esse método também chamamos de contrutor.
  #Esse método de inicialização possui parâmetros.
  #A variável self é usada para representar a instância da classe que é freqüentemente usada na programação orientada a objetos.
  def __init__(self,nome,login,senha):
    self.nome = nome
    self.login = login
    self.senha = senha

  def imprimeDados(self):
    print(f'nome:{self.nome}\nlogin:{self.login}\nsenha:{self.senha}')
```

**Instanciando um usuario**

```
#Criando uma instância da classe Usuário
user1 = Usuario("José Luiz",'deboas','123')
user1.imprimeDados()
```

```
#Criando outra instância da classe Usuário
user2 = Usuario("Fernanda",'Fer','321')
user2.imprimeDados()
```

```
#Acessando o atributo
user1.nome
```

```
user2.nome
```

**Exemplo da Classe Usuário 2º Exemplo. Em Python não há modificadores de acesso do tipo Private, Protected como em Java. Em python tudo é Público**

```
class Usuario():
  def __init__(self):
    self.nome = ""
    self.login = ""
    self.senha = ""

  def setNome(self, nome):
    self.nome = nome;

  def setLogin(self, login):
    self.login = login

  def setSenha(self, senha):
    self.senha = senha

  def getNome(self):
    return self.nome

  def getLogin(self):
    return self.login

  def getSenha(self):
    return self.senha  

  def imprimeDados(self):
    print(f'nome:{self.nome}\nlogin:{self.login}\nsenha:{self.senha}')
```

```
user3 = Usuario()
user3.setNome("José Luiz Vilas Boas")
user3.setLogin("deboas")
user3.setSenha("123")
user3.imprimeDados()
```

```
#Acessando o nome pelo método getNome
user3.getNome()
```

```
#Acessando o nome diretamente pelo atributo
user3.nome
```

**Trabalhando com métodos**

```
#Criando uma classe chamada círculo
class Circulo():
  #Note: o valor de PI fora dos métodos dará a opção de ser uma constante, pois o usuário não poderá alterá-lo.
  PI = 3.14

  #Criando um construtor 
  def __ini__(self):
    self.raio = 0

  def setRaio(self, raio):
    self.raio = raio

  def getRaio(self):
    return self.raio

  #calculando a áres
  def area(self):
    return  pow(self.raio,2) * Circulo.PI
```

```
#Criando o objeto
circulo = Circulo()
```

```
#Passando o valor para o raio
circulo.setRaio(5)
```

```
#imprimindo o raio
print(f'Raio = {circulo.getRaio()}')
```

```
#Imprimindo a área
print(f'Área = ', circulo.area())
```

**Herança**

```
#Criando a classe Pessoa
class Pessoa():

  def __init__(self):
    self.nome = ''   

  def setNome(self, nome):
    self.nome = nome;  

  def getNome(self):
    return self.nome

  def imprimeDados(self):
    print(f'nome = {self.nome}')
```

```
pessoa = Pessoa()
```

```
pessoa.setNome('José')
pessoa.imprimeDados()
```

```
#Criando a sub classe aluno
class Aluno(Pessoa):

  def __ini__(self):
    Pessoa.__init__(self)
    #super().__init__(self)
    self.matricula = ''
  
  def setMatricula(self, matricula):
    self.matricula = matricula;  

  def getMatricula(self):
    return self.matricula
   
  def imprimeDadosAluno(self):
    print(f'Nome = {self.nome}\nMatricula = {self.matricula}') 

```

```
aluno = Aluno()
```

```
aluno.setNome('Jose')
#chamado o método da super classe ou classe mãe
aluno.imprimeDados()
```

```
aluno.setMatricula(1234560)
#chamado o método da sudclass ou classe filha
aluno.imprimeDadosAluno()
```

```
print(f'Nome do aluno:{aluno.getNome()}')
```

```
print(f'Nome do aluno:{aluno.nome}')
```

**Polimorfismo**

```
class Carros():
  def __init__(self,nome,cor):
    self.nome = nome
    self.cor = cor
  def imprimeDadosCarros(self):
    print(f'nome = {self.nome}\ncor = {self.cor}') 
```

```
carros = Carros('Corsa','Branco')
carros.imprimeDadosCarros()
```

```
class Gol(Carros):
  def __init__(self, nome, cor):
    super().__init__(nome,cor)
  
  def imprimeDadosGol(self):
    print(f'nome = {self.nome}\ncor = {self.cor}') 
```

```
gol = Gol('Gol GTI','Prata')
gol.imprimeDadosGol()
```

**Métodos especiais : https://www.phylos.net/2021-05-25/python-classes-metodos-especiais/**

**Métodos especiais, ou métodos mágicos, em Python são métodos predefinidos em todos os objetos, com invocação automática sob circunstâncias especiais. Eles normalmente não são chamados diretamente pelo usuário mas podem ser overloaded (sobrescritos e alterados). Seus nomes começam e terminam com sublinhados duplos chamados de dunder (uma expressão derivada de double underscore).**

```
class Livro():
  def __init__(self,titulo, autor, paginas):
    print('Livro criado')
    self.titulo = titulo
    self.autor = autor
    self.paginas = paginas

  def __str__(self):
    return 'Títulos: %s, autor: %s , páginas: %s ' \
    %(self.titulo, self.autor, self.paginas)

  def __len__(self):
    return self.paginas

  def len(self):
    return print('Páginas do livro com método comum :', self.paginas)     
```

```
livro1 = Livro('Web Services - RESTFULL','Ricardo R. Lechete',431) 
```

```
#Quando chama o print, o método especial __str__ é chamado automaticamente
print(livro1)
```

```
#Chamando o __str__
str(livro1)
```

```
#Chamando o método especial len
len(livro1)
```

```
#Chamando o método len comum
livro1.len()
```

```
#Acessando um método que eu não criei na class
#__delattr__   , remover o atributo
del livro1.paginas
```
