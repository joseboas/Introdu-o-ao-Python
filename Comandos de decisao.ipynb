**Comando IF**

```
nota = 7.0
if nota >= 7.0:  
  print(f'Aprovado com nota: {nota}')
```

**if/else**

```
if nota >= 7.0:
  print(f'Aprovado com nota: {nota}')
else:
  print(f'Reprovado com nota: {nota}')
```

**elif**

```
#exibir qual número é maior, menor, caso contrário iguais
num1 = 9
num2 = 8
if num1 > num2:
  print('O primeiro número é maior')
elif num1 < num2:
  print('O segundo número é maior')
else:
  print('Os números são iguais')
```

**outros**

```
#exibir qual número é maior, menor, caso contrário iguais
num1 = 9
num2 = 9
if num1 > num2:  
  print('O primeiro número é maior')
if num1 < num2:
  print('O segundo número é maior')
if num1 == num2:
  print('Os números são iguais')
```

**Operadores Lógicos**

```
freq = 60
if (nota >= 7.0 and freq >= 75):
  print(f'Aprovado com média = {nota}')
else:
  print(f'Reprovado com média = {nota}')
```
