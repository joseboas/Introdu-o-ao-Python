**Função que retorna valor**

```
def fsoma(x,y):  
  z = x + y 
  return z
```

```
print(f'Soma de dois valores:{fsoma(3,5)}')
```

```
a = 2
b = 5
soma(a,b)
c = fsoma(a,b)
print(f'Soma de dois valores:{c}')
```

**Função com valor Default**

```
def intervalo(inic=1,fim=10):
  for inic in range(1, fim+1):
    print(inic)
```

```
intervalo()
```

**Passando vários argumntos quando não sei a quantidade ideal.**

```
def printVariavel(var1, *varios):
  print(f'Imprimindo o primeiro valor = {var1}')

  #imprimindo o valor do segundo elemento
  for i in varios:
    print(f'Imprindo os outros valores = {i}')
  
  return;
```

```
printVariavel(10)
```

```
printVariavel(10,20,30)
```
