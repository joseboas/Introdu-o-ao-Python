**Manipulação de Listas**

```
l1 = ['José','Maria','Pedro',33,'Pedro']
l1
```

**Verificando a qtde de um elemento na lista**

```
print(l1.count('Pedro'))
```

**Append: Permite adicionar elementos em uma lista no final**

```
l1.append(10)
l1
```

**Outra forma de adicionar**

```
#l1.extend([11,12])
l1
```

**Insert insere um novo elemento na posição desejada**

```
#Estou inserindo o numéro 9 posição desejada
l1.insert(1,9)
l1
```

**Remove: Remove um elementos de lista. Para remover um elemento é passar junto com o remove o elemento que deseja remover, por exemplo, remover da lista o Pedro.**

```
l1.remove('Pedro')
l1
```

**Del: Remove todos os elementos de uma lista de uma só vez. Contudo, após o comando ser executado, caso queira listar os elementos um erro será retornado, pois não existe mais elementos. O erro retornado é : name  <nomellista> is not defined.**

```
del(l1)
l1
```

**Copiando os elementos de uma lista para outro**

```

#cria uma lista vazia e outro com elemento
l2 = ['Jardim Alegre','Ivaiporã','Manuel Ribas']
l4= []
#Percorro l2 inserindo os elementos em l4
for item in l2:
  l4.append(item)
```

```
print(l4)
```

**Obtendo o index de um elemento**

```
l4.index('Jardim Alegre')
```

```
#Obtendo um indice de um item que não existe
l4.index('Guapirama')
```

**Ordenando listas**

```
l4.sort()
l4
```

**Lista reversa**

```
l4.reverse()
l4
```
