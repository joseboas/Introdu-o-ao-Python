**Exemplos**

```
profMineracaoDados = {'nome': 'José Luiz Vilas Boas','idade':43, 'formacoes':['Tecnólogo em Processamento de Dados','Especialista e Java', 
                                                                              'Mestre em Ciência da Computação']}                                                                             
```

```
profMineracaoDados 
```

**Imprimindo um dicionário**

```
print(profMineracaoDados)
```

**Pesquisa por uma chave específica**

```
profMineracaoDados['nome']
```

```
profMineracaoDados['formacoes']
```

**Adicionar novos elementos com novas chaves**

```
profMineracaoDados['trabalhoAtual'] = 'IFPR-Ivaiporã'
print(profMineracaoDados)
```

**Adicionar um valor a uma chave já existente**

```
profMineracaoDados['formacoes'].append('Analista de Dados em Python')
print(profMineracaoDados)
```

**A Atualização funciona da seguinte maneira: se você passar uma chave existente, o valor será atualizado com o novo, contudo, caso envie uma chave que não exista, a inserção será realizada. No exemplo abaixo passei a chave idade que já existe no dicionário, mas com um novo valor. Nesse caso, o valor será atualizado.**

```
profMineracaoDados['idade'] = 32
print(profMineracaoDados)
```

**Apagar um elemento**

```
del(profMineracaoDados)['formacoes']
print(profMineracaoDados)
```

**Observação: Caso eu execute apenas o comando del(profMineracaoDados), a variável toda ou o meu dicionário todo, será excluído**

**Visualização do dicionário completo, no formato dicionário, usa-se o função items().**

```
profMineracaoDados.items()
```

**Lista apenas as chaves**

```
profMineracaoDados.keys()
```

**Listar apenas os valores**

```
profMineracaoDados.values()
```

**Concatenar dois dicionários**

```
#Crio um novo dicionário
profMineracaoDados1 = {'formacoes': ['Tecnólogo em Processamento de Dados', 'Especialista e Java', 'Mestre em Ciência da Computação']}
print(profMineracaoDados1)
```

```
#Vou atualizar o profMineracaoDados com os dados do profMineracaoDados1
profMineracaoDados.update(profMineracaoDados1)
print(profMineracaoDados)
```

**Percorrer um dicionário**

```
for chave, valor in profMineracaoDados.items():
  print(f'Chave: {chave}, valor: {valor}') 
```

```
#Listando as chaves de um dicionário
for chave in profMineracaoDados:
  print(chave)
```

```
#Listando apenas os valores
for valor in profMineracaoDados.values():
  print(valor)
```

```
#Imprimindo a chave e um valor de um dicionário. O método items() retorna o valor
for chave,item in profMineracaoDados.items():
  print(chave,item)
```

**Criando um dicionário à partir de duas listas. Iremos utilizar a função ZIP.**

```
#Esta vai ser a chave
lista1 = ['A','B','C','D']
#Esta será o valor
lista2 = [1,2,3,4]
dic3 = dict(zip(lista1,lista2))
print(dic3)
```
