**List Comprehension**

```python
#a primeira variável (item) é o retorno. O mesmo nome tem que repetir, exemplo:item for item.
lista1 = [item for item in 'Python Básico']
```

```python
lista1
```

```python
lista2 = [x for x in range(10)]
```

```python
lista2
```

```python
#Elevar ao quadrado de 0 até 10
lista3 = [n**2 for n in range(11)]
```

```python
lista3
```

```python
#Criar uma  lista com os números pares de 0 a 10
lista4 = [n for n in range(11) if n % 2 == 0]
```

```python
lista4
```

```python
#Calcular a raiz quadrado dos núemeros da lista

raiz = [4,16,25,36]
r = [math.sqrt(x) for x in raiz]
```
