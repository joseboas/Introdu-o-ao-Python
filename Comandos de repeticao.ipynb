**Comando while. Enquanto cont for < ou igual a 5, o programa irá continuar a imprimir a variável. Na linha onde está o cont+=1, incrementa o valor de cont.Caso não faça isso a variável sempre irá valer 1, portanto, a condição nunca será alcançada e o programa entrará em um looping infinito.**

```
cont = 1
while cont <= 5:
  print(cont)
  cont +=1 
```

**While com True**

```
a = True
while a == True:
  print(a)
  if a == True:
    a = False
  print(a)  
```

**While com True outro modo**

```
a = True
while a:
  print(a)
  if a:
    a = False
  print(a)  
```

**Encerrando o While com else**

```
cont = 1
while (cont <= 5):
  print(cont)
  cont+=1 
else:
  print('Looping concluído')  
```

**Comando For. O for tem a variável i que controle o looping. O range define o intervalo a percorrer, no caso, será de (0,10). Isso significa que i irá iniciar com 0 irá parar quando chegar ao valor 9. Não precisa incrementar o valor de i como no while, pois o prório comando for já o faz isso**

```
for i in range(0,10):
  print(i)
```

**Range range(2, 10, 2) -> (inicio, parada, incremento). O Incremento pode ser negativo.**

```
#incremento positivo
for i in range(2,10,2):
  print(i)
```

```
#incremento negativo ou decremento
for i in range(10,0,-2):
  print(i)
```

```
#Looping aninhado
for i in range (0,5):
  for j in range (0,5):
    print(i,j)
```

**Comando break, continue e pass - Mudando o comportamento dentro do looping**

```
for i in range (0,10):
  print(i)
  if i == 5:
    break
```

**Comando continue. No exemplo abaixo ele irá pular o 5, mas irá continuar o looping**

```
for j in range (0,10):  
  if j == 5:    
    continue    #não executo nada, pulo a interação
  print(j) 
```

**Comando pass : pass que ocorre após a instrução condicional if está dizendo ao programa para continuar executando o loop e ignorar o fato de que a variável number é avaliada como equivalente a 5 durante uma das iterações, praticamente não faz nada**

```
for j in range (0,10):  
  if j == 5:
    pass  
  print(j) 
    
```
