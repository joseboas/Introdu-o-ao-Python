**Tratamento de Erros:Try, Except**

```
# Aqui eu espero um numero, mas se eu digitar um caractere 
try:
  a = int(input("Digite um número:"))
except:
  print("Valor inválido!!!")
else:
  print(f"Número digitado {a}")
```

```
#incrementando o código anterior
while True:
  try:
    a = int(input("Digite um número:"))
  except:
    print("Valor inválido!!!")
  else:
    print(f"Número digitado {a}")
    break 
```

```
#incrementando ainda mais
while True:
  try:
    a = int(input("Digite um número:"))
  except ValueError:
    print("Valor inválido!!!")
  except KeyboardInterrupt:
    print(f"Usuário parou a execução ")
    break
  else:  
    print(f"Número digitado {a}")
    break 
 #quando clico no stop o programa vai para o  KeyboardInterrupt:  
```

```
#Criando um arquivo
try:
  f = open('/content/arquivo.txt','w')
except IOError:
  print('Arquivo não pode ser criado')
else:
  print('Arquivo criado com sucesso ')
  f.close 
```

```
#Verifcando um arquivo
try:
  f = open('/content/arquivos.txt','r')
except IOError:
  print('Arquivo não encontrado')
else:
  print('Arquivo encontntrado')   
```

**Finnaly**

```
#Tudo que estiver dentro do finnaly será executado de qualquer modo.
#Verifcando um arquivo
try:
  f = open('/content/arquivos.txt','r')
except IOError:
  print('Arquivo não encontrado')
else:
  print('Arquivo encontrado')  
finally:
  f.close  
  print('Finnaly sempre será executado')
```
