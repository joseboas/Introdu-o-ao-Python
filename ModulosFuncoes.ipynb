**Funções abs, round, pow**

```
#abs = retorno o valor absoluto,tira o valor negativo
print(abs(-8))
print(abs(-8))
```

```
#round = aredonda
a = 3.14159 
print(round(a,2))
print(round(a,3))
print(round(a,1))
```

```
#pow = potência
print(pow(5,2))
print(pow(5,3))
```

**Funções prontas. Importando módulos ou bibliotecas**

```
import statistics
```

```
#mean retorna a média
z = [4,8,9,5,6,4.7,20]
x = statistics.mean(z)
print(x)
```

```
#median retorna a mediana ou valor do meio
x = statistics.median(z)
print(x)
```

**Usando nome alternativo.**

```
import statistics  as est
z = [4,8,9,5,6,4.7,20]
x = est.mean(z)
print(x)
```

```
x = est.median(z)
print(x)
```

**Uso sem nome da módulo, mas importando funções específicas**

```
x = median(z)
print(x)
```
