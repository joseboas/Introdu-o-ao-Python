**Estrutura de decisão com and e or**

**Utilizando o and**

```
nota1 = float(input("Digite a nota do primeiro bimestre:"))
nota2 = float(input("Digite a nota do segundo bimestre:"))
media = (nota1 + nota2) / 2
freq = int(input("Digite a frequência:"))
if (media >= 7 and freq >= 75):
  print(f'Aprovado com média = {media} e frequência = {freq}')
else:
  print(f'Reprovado com média = {media} e frequência = {freq}')
```

**Utilizando o or**

```
nota1 = float(input("Digite a nota do primeiro bimestre:"))
nota2 = float(input("Digite a nota do segundo bimestre:"))
media = (nota1 + nota2) / 2
freq = int(input("Digite a frequência:"))
if (media >= 7 or freq >= 75):
  print(f'Aprovado com média = {media} e frequência = {freq}')
else:
  print(f'Reprovado com média = {media} e frequência = {freq}')
```

**Usando com o elif**

```
nota1 = float(input("Digite a nota do primeiro bimestre:"))
nota2 = float(input("Digite a nota do segundo bimestre:"))
media = (nota1 + nota2) / 2
if media <= 4:
  print(f'Reprovado com média = {media}') 
elif media > 4 and media < 7:
  print(f'Exame com media = {media}')  
else:
  print(f'Aprovado com média = {media}')
```

**if aninhado**

```
nota1 = float(input("Digite a nota do primeiro bimestre:"))
nota2 = float(input("Digite a nota do segundo bimestre:"))
media = (nota1 + nota2) / 2
freq = int(input("Digite a frequência:"))
if media >= 7:
  if  freq >= 75:
    print(f'Aprovado com média = {media} e frequência = {freq}')
  else:
    print(f'Reprovado com média = {media} e frequência = {freq}')
else:
  print(f'Reprovado com média = {media}')
```
