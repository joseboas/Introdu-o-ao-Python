**União: Os dois conjuntos serão unidos e os elementos iguais que estão em ambos os conjuntos irão aparecer apenas uma vez na nova relação.**

```
c1 = {1,2,3}
c2 = {3,4,5}
c1.union(c2)
```

**Interseção: Lista os elementos comuns nos dois conjuntos.**

```
c1.intersection(c2)
```

**Diferença entre os conjuntos: Lista os elementos que estão no primeiro conjunto comparado que não estão no segundo**

```
c1.difference(c2)
```

```
#Fazer o contrário
c2.difference(c1)
```
