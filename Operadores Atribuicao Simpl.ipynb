**Operadores**

```
a = 5
b = 6
a += 6
print("Valor de a =",a)
```

```
 a -= b
 print("Valor de a =",a)
```

```
a *= 2
print("Valor de a =",a)
```

```
a /= 2
print("Valor de a =",a)
```
