**Manipulação de Arquivos:** 

*   open()      -> usada para abrir arquivo 
*   read()      -> leitura de arquivo
*   write()     -> gravação no arquivo 
*   seek()      -> retorna para o início do arquivo
*   readlines() -> retorna a lista de linhas do arquivo
*   close()     -> fecha o arquivo

**Manipulando Arquivo Texto**

```
#Abrindo arquivo em modo leitura
arq1 = open('/content/arquivo1.txt','r')
```

```
#Lendo arquivo
print(arq1.read())
```

```
#Contar o número de caracteres
print(arq1.tell())
```

```
#Retorna para o início do arquivo. O comando abaixo irá retorna a "POSIÇÃO" da primeira linha  e coluna
print(arq1.seek(0,0))
```

```
#Ler os 15 primeiros caracteres. Na verdade ele lê a partir que o seek utilizou
print(arq1.read(15))
```

```
#Mudando o seek
#Vou copiar a partir da coluna 5
print(arq1.seek(5))
print(arq1.read(15))
```

**Note: o \n indica uma frase em cada linha.**

**Escrever um aquivo de texto**

```
#Utilizo o w, pois eu irei gravar
arq2 = open('/content/arquivo1.txt', 'w')
```

```
#Aqui irá dar um erro, pois eu não abri para leitura
print(arq2.read())
```

```
#Escrevendo no arquivo
arq2.write("Verificar se irá gravar no arquivo")
```

```
#Fechando o arquivo
arq2.close()
```

```
#Lendo o arquivo gravado
arq2 = open('/content/arquivo1.txt','r')
```

```
print(arq2.read())
```

**Acrescentando conteúdo. Abrindo no módulo append**
