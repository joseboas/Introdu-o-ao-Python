```python
import time
```

```python
#Otendo os segundos
seg = time.time()
print(seg)
```

```python
#hora local
horaLocal = time.ctime(seg)
print("Hora local:", horaLocal)
```

```python
#sleep :função suspende (atrasa) a execução da thread atual por um determinado número de segundos.
print('Imediatamente')
time.sleep(2)
print('Depois de 2 segundos ')
```

```python
#Cálculo entre segundos
print('Inicio')
inicio = time.time()
time.sleep(2)
fim = time.time()
tempo = (fim - inicio) 
print(f'Tempo de execução :{tempo}')
```

```python
#Exibindo um a um
result = time.localtime(1545925769)
print("result:", result)
print("\nAno:", result.tm_year)
print("Hora:", result.tm_hour)
```

```python
#Atribuindo o a localtime manualmente
t = (2021, 12, 28, 8, 44, 4, 4, 362, 0)
horaData = time.mktime(t)
print("Hora Local:", horaData)
result = time.localtime(horaData)
print("\nAno:", result.tm_year)
print("Hora:", result.tm_hour)
```

```python
#Atribuindo o a localtime atual manualmente
#Lembrando que eu pego os dados dos servidor Google.
result = time.localtime(time.time())
print("result:", result)
print("Dia:", result.tm_mday)
print("Mês:", result.tm_mon)
print("Ano:", result.tm_year)
print("Hora:", result.tm_hour)
print("Segundo:", result.tm_sec)
```
