**Exemplo de Listas**

**Somente números inteiros**

```
lista1 = [1,2,3,4,5,6,7]
print(lista1)
```

**Números inteiros e reais**

```
lista2 = [1,2,6.5,1.2]
print(lista2)
```

**Lista com numeros, textos e valores lógicos**

```
lista3 = [1,True,'False,3.5',3.5]
print(lista3)
```

**Lista dentro de outra lista ou listas aninhadas**

```
lista4 = [1,2,[3,4,5,'False'],6,7,True]
print(lista4)
```

**Criando uma lista com range**

```
lista5 = list(range(0,10)) 
lista5
```

```
print(len(lista4))
```

**Acessando elementos**

```
print(lista4[0])
```

```
print(lista4[2])
```

```
print(lista4[5])
```

```
lista5 = lista4[2]
lista5
```

```
lista5[2]
```

```
lista4[2][2]
```

**Operações**

```
a = lista4[2][2] * 10
a
```

```
b = lista4[2][1] + 3
b
```

**Outra forma de exibir listas**

```
print(len(lista4))
lista4[0:len(lista4)]
```

```
lista4[0:5]
```

**Concatenando listas**

```
lst1 = [1,2,3,4]
lst1
```

```
lst2 = [5,6]
lst2
```

```
concat = lst1 + lst2
concat
```

**Operador in**

```
#verificar se tem um elemento na lista
print(5 in concat)
```

```
print(10 in  concat)
```

**Percorrendo uma lista**

```
for i in range(0,len(lista4)):
  print(lista4[i])
```

```
lista5 = ['Introdução a Computação','Engenharia de Software','Estrutura de Dados']
for elemento in lista5:
  print(elemento)
```

**Funções Built-in**

```
maior = max(concat)
maior
```

```
#Obtendo menor elementro da lista
menor = min(concat)
menor
```

```
#Obtendo  tamanho de registro
tam = len(concat)
tam
```

**Extras**

```
lst6 = [1,2,3]
lst7 = [4,5,6]
for i in lst6:
  for j in lst7:
    print(i,j)
```

```
lst6 = [[1,2,3],[4,5,6]]
for i in lst6:  
    print(i)


```
